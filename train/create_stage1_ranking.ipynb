```python
import json
import numpy as np
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
trainval = json.load(open("/home/uyen/workspace/nlp_project/data/e2e_qa/e2eqa-train+public_test-v1/zac2022_train_merged_final.json"))
mailong25 = json.load(open("/home/uyen/workspace/nlp_project/repos/bert-vietnamese-question-answering/dataset/train_IR.json"))
```

```python
full_anno = [x for x in trainval['data'] if x['category'] == "FULL_ANNOTATION"]
train, val = train_test_split(full_anno, random_state=42, test_size=0.15)
full_anno = dict({
    'data': val
})
for x in full_anno['data']:
    x['label'] = x['answer']
    x['answer'] = None
```

```python
val_ids = [x['id'] for x in val]
```

```python
texts = []
labels = []
questions = []
for x in trainval['data']:
    if x['id'] in val_ids:
        continue
    # texts.append(x['dirty_text'] or x['text'])
    texts.append(x['text'])
    questions.append(x['question'])
    if x['category'] == 'FULL_ANNOTATION'or x['is_long_answer']:
        labels.append(1)
    else:
        labels.append(0)

for item in mailong25:
    texts.append(item['text'])
    questions.append(item['question'])
    if item['label']:
        labels.append(1)
    else:
        labels.append(0)

df = pd.DataFrame()
df["text"] = texts
df["question"] = questions
df["label"] = labels

print(len(df))
```
